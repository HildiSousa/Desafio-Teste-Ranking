```python
import pandas as pd
from openpyxl import load_workbook

df_dados = ['/content/Ranking_2023.xlsx']  # Apenas o último arquivo está sendo processado

arquivos_coordenadas = {
    '/content/Ranking_2023.xlsx': {'AS': (2, 32, 5, 30)}  # Apenas as coordenadas para o último arquivo
}

tabelas = []

for df_dado in df_dados:
    wb = load_workbook(df_dado, data_only=True)
    coordenadas = arquivos_coordenadas[df_dado]
    inicio_linha, fim_linha, inicio_coluna, fim_coluna = coordenadas['AS']
    ws = wb['AS']
    dados_tabela = []
    for row in ws.iter_rows(min_row=inicio_linha, max_row=fim_linha, min_col=inicio_coluna, max_col=fim_coluna):
        linha = [cell.value for cell in row]
        dados_tabela.append(linha)
    df_temp = pd.DataFrame(dados_tabela)
    tabelas.append(df_temp)

# Extrair as colunas 0, 1 e a última coluna de cada DataFrame na lista tabelas
colunas_extraidas = []
for df_temp in tabelas:
    coluna_0 = df_temp.iloc[:, 0]  # Primeira coluna
    coluna_1 = df_temp.iloc[:, 1]  # Segunda coluna
    ultima_coluna = df_temp.iloc[:, -1]  # Última coluna
    colunas_extraidas.append(pd.concat([coluna_0, coluna_1, ultima_coluna], axis=1))

# Criar um novo DataFrame com as colunas extraídas
novo_dataset = pd.concat(colunas_extraidas)

# Excluir as linhas 1 e 3 do novo DataFrame
novo_dataset = novo_dataset.drop(index=[1, 3])

# Salvar o dataset em um arquivo Excel
nome_arquivo = 'novo_dataset.xlsx'  # Nome do arquivo
novo_dataset.to_excel(nome_arquivo, index=False)
print(f'O arquivo {nome_arquivo} foi salvo com sucesso!')
# Mostrar as primeiras linhas do novo dataset
print(novo_dataset.head())





```
