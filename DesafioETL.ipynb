```python
#Importando Bibliotecas

import pandas as pd
import numpy as np
from openpyxl import load_workbook
```

```python
#Importando os arquivos Excel
df_dados = ['/content/Ranking_2016.xlsx', '/content/Ranking_2017.xlsx', '/content/Ranking_2018.xlsx',
            '/content/Ranking_2019.xlsx', '/content/Ranking_2020.xlsx', '/content/Ranking_2021.xlsx',
            '/content/Ranking_2022.xlsx', '/content/Ranking_2023.xlsx']
```

```python
# Criando Dicionários com as tabelas Excel

arquivos_coordenadas = {
    '/content/Ranking_2016.xlsx': {'AS': (2, 32, 5, 23)},
    '/content/Ranking_2017.xlsx': {'AS': (2, 32, 5, 24)},
    '/content/Ranking_2018.xlsx': {'AS': (2, 32, 5, 25)},
    '/content/Ranking_2019.xlsx': {'AS': (2, 32, 5, 26)},
    '/content/Ranking_2020.xlsx': {'AS': (2, 32, 5, 27)},
    '/content/Ranking_2021.xlsx': {'AS': (2, 32, 5, 28)},
    '/content/Ranking_2022.xlsx': {'AS': (2, 32, 5, 29)},
    '/content/Ranking_2023.xlsx': {'AS': (2, 32, 5, 30)}


}
```

```python
# Criando lista para armazenar os bases de dados
tabelas = []
```

```python
 #Processando os arquivos Excel

for df_dado in df_dados:
  pass
```

```python
#Carregando o arquivo
wb = load_workbook(df_dado, data_only=True)
```

```python
# Extraindo as coordenadas da tabela para outro arquivoarquivo
coordenadas = arquivos_coordenadas[df_dado]
```

```python
# Selecionando as coordenadas
inicio_linha, fim_linha, inicio_coluna, fim_coluna = coordenadas['AS']
```

```python
  # Selecionando a planilha com o dados
ws = wb['AS']
```

```python
 # Extraindo os dados da tabela
dados_tabela = []
for row in ws.iter_rows(min_row=inicio_linha, max_row=fim_linha, min_col=inicio_coluna, max_col=fim_coluna):
        linha = [cell.value for cell in row]
        dados_tabela.append(linha)
```

```python
# Convertendo os dados em DataFrame e adicionando à lista de tabelas
df_temp = pd.DataFrame(dados_tabela)
tabelas.append(df_temp)
```

```python
# Combinando todos os dados em um único Base
df_final = pd.concat(tabelas, ignore_index=True)

# Exibindo os primeiros registros do Base final
print(df_final.head())
```

```python
df_final.head()
```

```python
# Extraindo apenas as colunas de 2015 a 2023 da base de dados final
df_final_2015_2023 = df_final.iloc[:, [0,1] + list(range(17, 26))]
```

```python
# Exibindo os primeiros registros com colunas de 2015 a 2023
print(df_final_2015_2023.head())
```

```python
#Mostrando os dados com a extração das colunas desejadas

df_final_2015_2023.head()
```

```python
# Excluindo as linhas 1 e 3 da base de dados
df_unificado = df_final_2015_2023.drop([1, 3])
```
